# IV. Data Cleaning

11. (3б) Перетворити колонку launched у цілочислену (Int64). Перевірити, чи все пройшло, як очікувалось.

```python
df['launched'] = df['launched'].fillna(0).astype('int64')
print(df.dtypes)
```

12. (3б) Видалити колонки, де кількість непустих значень становить менше 30% від загальної кількості записів. Можна вручну порахувати і тут тільки видалити. За автоматизацію розрахунку відстоків - додаткові 3 бали.

```python
df=df.drop(columns=[col for col in list(df.columns.values) if (df[col].notnull().count()<df[col].count()*.3)])
for col in df:
    print(f"Назва колонки: {col}\tТип: {df[col].dtype}\tКількість непустих значень:{df[col].notnull().sum()}")
```

13. (4б) Очистити датафрейм від дублікатів (дублюючихся рядків) та записів, де атака продовжувалась відʼємну або нульову кількість часу (що фізично неможливо). Не забудьте оновити індекс. Скільки рядків було видалено? Як ви це перевірили у коді?

```python
original_rows=len(df)
df=df.drop_duplicates()
df=df.drop(index=[x for x in df.index if pd.Timedelta(pd.to_datetime(df["time_end"][x])-pd.to_datetime(df["time_start"][x])).seconds<=0])
```

15. (9б) Зʼясуйте, який середній відсоток збиття ракет за весь наведений період.

```python
d=[(x/y) for (x,y) in zip(df["destroyed"],df["launched"]) if y!=0]
print(f"Середній відсоток збиття: {100*sum(d)/len(d)}%")
```

16. (11б) Зʼясуйте 10 найпопулярніших моделей запущених пристроїв за кількістю випущених одиниць за наведений період. Запишіть результат у файл типу json. Скопіюйте зміст файлу та вставте нижче.

```python
models=dict([(key,0) for key in df['model'].unique()])
for i in df.index:
    models[df['model'][i]]+=int(df['launched'][i])
models=dict(sorted(models.items(),key=lambda x:x[1])[::-1][:10])
import json
with open('models.json', 'w') as file:
    json.dump(models, file)
print(open('models.json','r').read())
```

17. (10б) Зʼясуйте, у який день було найбільше влучань за весь період (незбитих пристроїв). Виведіть час початку атаки та кількість влучань.

```python
max_i,max_n=0,0
for i in df.index:
    if df['launched'][i]-df['destroyed'][i]>max_n:
        max_i,max_n=i,df['launched'][i]-df['destroyed'][i]
print(f"Найбільше влучань - {max_n} під час атаки {df['time_start'][max_i]}")
```

18. (15б) Знайдіть найбільш тривалу та найбільш коротку атаки за кожен представлений місяць. Результат отримайте у такому вигляді:

![текст посилання](https://drive.google.com/uc?id=1gISSxR2mXjTEOgoAk_vQnXJeqoL-eA5b)


Запишіть його у файл типу сsv. Після цього виведіть на екран.

```python
start_date = pd.to_datetime(str(df['time_start'].min())[:10])
end_date = pd.to_datetime(str(df['time_end'].max())[:10])
all_months = list(map(lambda x:str(x)[:7],list(pd.date_range(start=start_date, end=end_date, freq='ME'))))
months=dict([(x,[0,'23:00:00']) for x in all_months])
for i in df.index:
    if pd.Timedelta(pd.to_datetime(df['time_end'][i])-pd.to_datetime(df['time_start'][i])).seconds>pd.Timedelta(months[df['time_start'][i][:7]][0]).seconds:
        months[df['time_start'][i][:7]][0]=str(pd.to_datetime(df['time_end'][i])-pd.to_datetime(df['time_start'][i]))
    if pd.Timedelta(pd.to_datetime(df['time_end'][i])-pd.to_datetime(df['time_start'][i])).seconds<pd.Timedelta(months[df['time_start'][i][:7]][1]).seconds:
        months[df['time_start'][i][:7]][1]=str(pd.to_datetime(df['time_end'][i])-pd.to_datetime(df['time_start'][i]))
fields=['year','month','longest_attack_duration','shortest_attack_duration']
rows=[[m[:4],m[5:],d[0],d[1]] for (m,d) in months.items()]
import csv
with open('most.csv','w') as file:
    c=csv.writer(file)
    c.writerow(fields)
    c.writerows(rows)
print(pd.read_csv('most.csv'))
```
